Example for the definition of the independent and dependent variables:

## 2.1. Regression model with multiple variables and with a constant for Situation 1 (repeat for all the situations)

```python
x1wx= df[['CRIM','ZN','INDUS']]
y = target["MEDV"]
x1wx = sm.add_constant(x1wx)
model1wx = sm.OLS(y,x1wx).fit()
model1wx.summary()
```

## 2.1. Regression model with multiple variables and with a constant for Situation 2

```python
x2wx= df[['CHAS','NOX','RM']]
y = target["MEDV"]
x2wx = sm.add_constant(x2wx)
model2wx = sm.OLS(y,x2wx).fit()
model2wx.summary()
```

## 2.1. Regression model with multiple variables and with a constant for Situation 3

```python
x3wx= df[['AGE','DIS','RAD']]
y = target["MEDV"]
x3wx = sm.add_constant(x3wx)
model3wx = sm.OLS(y,x3wx).fit()
model3wx.summary()
```

## 2.1. Regression model with multiple variables and with a constant for Situation 4

```python
x4wx= df[['TAX','PTRATIO','B','LSTAT']]
y = target["MEDV"]
x4wx = sm.add_constant(x4wx)
model4wx = sm.OLS(y,x4wx).fit()
model4wx.summary()
```

## 2.1. Regression model with multiple variables and with a constant for Situation 5

```python
x5wx= df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
y = target["MEDV"]
x5wx = sm.add_constant(x5wx)
model5wx = sm.OLS(y,x5wx).fit()
model5wx.summary()
```

### Interpreting the fitness of the models

Models without x tend to work better than with their x counterparts, although for actually modeling the real issue it should be better to always use the with x variants. The best model with x would be model 5 with an rsquared of .741, although in some cases like INDUS and AGE the P value is to high as well as the std error of NOX is to high, a model without this 3 would be, in theory better.
