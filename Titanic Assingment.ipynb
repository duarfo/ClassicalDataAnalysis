```python
# imports
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=df.dropna()
```

```python
df1.head()
```

```python
X = df1[["Age", "Fare"]]
y = df1["Survived"]
```

```python
X = sm.add_constant(X) 
X.head()
```

```python
model = sm.OLS(y, X).fit()
predictions = model.predict(X) 
predictions
```
